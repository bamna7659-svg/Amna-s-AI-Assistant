```python
# Agar library install nahi hai to ye line chalaein
!pip install -U groq

from groq import Groq
from google.colab import userdata

# Yahan se apni key uthayein (Secret ka naam GROQ_API_KEY hi hona chahiye)
api_key = userdata.get('GROQ_API_KEY')
client = Groq(api_key=api_key)

def chat_with_memory():
    print("Groq AI Bot is ready! (Type 'quit' to exit)")
    # Memory ke liye history list
    messages = [{"role": "system", "content": "You are a helpful assistant for Amna who manages her YouTube AI channel and business."}]

    while True:
        user_input = input("You: ")
        if user_input.lower() == "quit":
            print("Goodbye!")
            break

        if not user_input.strip():
            continue

        messages.append({"role": "user", "content": user_input})

        try:
            # Updated model name jo abhi active hai
            chat_completion = client.chat.completions.create(
                messages=messages,
                model="llama-3.3-70b-versatile",
            )

            reply = chat_completion.choices[0].message.content
            print(f"Bot: {reply}")

            # Bot ka jawab history mein add karein taake wo yaad rakhe
            messages.append({"role": "assistant", "content": reply})

        except Exception as e:
            print(f"Error: {e}")

# Function run karein
chat_with_memory()
```
